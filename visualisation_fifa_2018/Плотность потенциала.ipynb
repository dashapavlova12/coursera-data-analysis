Текстовое описание : плотность распределения потенциала игрока от возраста показывает что наибольшим потенциалом обладают молодые игроки (20-25 лет), с возрастом потенциал игрока падает 

Плотность распределения потенциала игрока от возраста укажет на возрастную категорию игроков с наибольшим потенциалом. Также можно определить в какой категории потенциал больше среднего а в какой меньше

```
fig, ax = plt.subplots(figsize = (20,10))
df.Potential.plot.kde(label = "All", ax = ax)
age1_df = df[np.logical_and(df.Age >= 16, df.Age < 20)]
age2_df = df[np.logical_and(df.Age >= 20, df.Age <= 25)]
age3_df = df[np.logical_and(df.Age > 25, df.Age <= 30)]
age4_df = df[np.logical_and(df.Age > 30, df.Age <= 35)]
age5_df = df[np.logical_and(df.Age > 35, df.Age <= 41)]
age1_df.Potential.plot.kde(label = '16-19', ax = ax)
age2_df.Potential.plot.kde(label = '20-25', ax = ax)
age3_df.Potential.plot.kde(label = '26-30', ax = ax)
age4_df.Potential.plot.kde(label = '31-35', ax = ax)
age5_df.Potential.plot.kde(label = '36-41', ax = ax)
ax.set_title("Potential density distribution", fontdict = {'fontsize' : 15})
plt.legend()
plt.xlim(50,90)
plt.xlabel('Potential')
```
