Распределение зарплат в зависимости от возраста в клубах Real Madrid и Barcelona показывает уровень зарплат каждой возрастной группы 

```
fifa_fig = px.scatter(df[np.logical_or(df.Club == 'Real Madrid CF',df.Club == 'FC Barcelona')], x = 'Age', y = 'Wage', color = 'Club')
fifa_fig.update_layout(title = 'Распределение зарплат в зависимости от возраста в клубах Real Madrid и Barcelona')
fifa_fig.show()
```
